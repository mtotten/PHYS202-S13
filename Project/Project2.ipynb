```python
import numpy.random
#define the road; length and number of cars
def Road(L,N):
    #set velocities to array of zeros with length N, number of cars
    velocities = zeros(N,int)
    #set an array of zeros with length N for cycling back to start
    loopreturn = zeros(N,int)
    #variable CarMinusOne for count (car behind)
    for CarMinusOne in range(0,N):
        #assign random velocities for cars on the road
        velocities[CarMinusOne] = randint(0, L-N)
    #set cars as an array of zeros with length N, number of cars
    cars = zeros(N,int)
    #cycle through the car positions on the road
    for i in range(N):
        cars[i] = i + 1
    #set road as and array of zeros with length L
    road = zeros(L,int)
    #car in second position
    car = 1
    #set position as an array of zeros with lengh N, number of cars
    position = zeros(N,int)
    #amount of cars must not fill all spaces in the road
    while sum(road) != sum(cars):
        #assign a random position between 0 and L, length of the road
        i = randint(0,L)
        if road[i] == 0:
            road[i] = car
            position[car-1] = i
            car +=1
    return position, cars, road, velocities, loopreturn
#find the number of empty slots in front of each car
def GetGSubI(road,car,position):
    #define L as length of the road
    L = len(road)
    #define position of car (car-1+1=car)
    x = position[car-1] + 1
    #position must be less than L for the car to be on the road
    while x < L:
        #if car is in first slot, the car does not move during time step
        if road[x] != 0:
            break
        #count up by one car if car is in first slot to get to second slot
        x += 1    
    #number of empty slots is position of car - position of car in front - 1 (-1 to account for inclusivity)
    g = x - position[car-1] - 1
    #if car is in last slot, cycle back to the first car, index 0
    if x == L:
        x = 0
        #when car is in last slot, amount of slots in front is the position of the car + position of the car in front - 1 (for index)
        while road[x] == 0:
            x += 1    
        g = x + L - position[car-1] - 1
    #get the number of slots in front of car, g
    return g
#define "stochastic traffic cellular automata" or SCTA
#input the length of the road, number of cars, position, velocity and probability of each, loopreturn for last car
def STCA(road,cars,position,velocities,probability,loopreturn):
    # N is the number of cars
    N = len(cars)
    # L is the length of the road
    L = len(road)
    #for car behind 
    for CarMinusOne in range(0,N):
        #find the number of spaces in front of the car (car = CarMinusOne +1)
        g = GetGSubI(road,CarMinusOne + 1,position)
        #if velocity is greater than spaces in front of car, make velocity equal to number of spaces (v=g)
        if velocities[CarMinusOne] > g:
            velocities[CarMinusOne] = g
        #if velocity is less than spaces in front of car, increase the velocity by one unit
        elif velocities[CarMinusOne] < g:
            velocities[CarMinusOne] += 1
        #for a velovity greater than 0, allow for random slowdowns (decrease velocity by 1 unit) with probability p
        if rand() <= probability and velocities[CarMinusOne] > 0:
            velocities[CarMinusOne] -= 1     
    #move all cars forward by velocity
    for CarMinusOne in range(0,N):
        position[CarMinusOne] += velocities[CarMinusOne]
        if position[CarMinusOne] >= L:
            #last car increase by 1 to return to front
            loopreturn[CarMinusOne] += 1 
            #position back to L
            position[CarMinusOne] -= L
    #set road equal to array of zeros with length L
    road = zeros(L,int)
    #car position on the road = CarMinusOne +1, cycle through cars in loop 
    for CarMinusOne in range(0,N):
        road[position[CarMinusOne]] = CarMinusOne + 1 
    #return parameters
    return road, cars, position, velocities, loopreturn        
#define "asymmetric stochastic exclusion principle" or ASEP  
#input the length of the road, number of cars, position and velocity of each, loopreturn for last car
def ASEP(road,cars,position,velocities,loopreturn):
    #N is the number of cars
    N = len(cars)
    #L is the length of the road
    L = len(road)
    #pick one car at random
    Car = randint(0,N)
    #find number of slots in front of car, g
    g = GetGSubI(road,Car,position)
    
    
    x = position[Car-1] + 1
    
    while x < L:
        #if car is in first slot, the car does not move during time step
        if road[x] != 0:
            break
        #count up by one car if car is in first slot to get to second slot
    x += 1
    
    
    #if the velocity of the car is greater than the number of spaces in front, set the velocity equal to g, the number of slots in front
    if velocities[Car-1] > g:
        velocities[Car-1] = g
    #if the velocity of the car is less than the number of spaces in front, increase the velocity by one unit    
    elif velocities[Car-1] < g:
        velocities[Car-1] += 1
    #move this car position forward by the velocity    
    position[Car - 1] += velocities[Car -1]
    #account for a car at the end of the line; return the count to the front
    if position[Car -1] >= L:
        position[Car -1] -= L
        loopreturn[Car - 1] += 1
    road = zeros(L,int)
    #loop through all cars to get position and velocity for each
    for CarMinusOne in range(0,N):
        road[position[CarMinusOne]] = CarMinusOne + 1
    return road, cars, position, velocities, loopreturn             
            
    
```

```python
def RunTimeSTCA(L,N,Runs):
    P,C,R,V,LR = Road(L,N)
    Vavg = zeros(Runs, float)
    Positions = zeros((Runs,len(P)), int )
    Current = zeros(Runs, float)
    Pr = .3
    for i in range(Runs):
        R,C,P,V,LR = STCA(R,C,P,V,Pr,LR)
        Current[i] = float(sum(V)) / float(Runs)
        Vavg[i] = float(sum(V)) / float(len(V))
        Positions[i] = P+L*LR
    return Current, Vavg, Positions

def RunTimeASEP(L,N,Runs):
    P,C,R,V,LR = Road(L,N)
    Vavg = zeros(Runs, float)
    Positions = zeros((Runs,len(P)), int )
    Current = zeros(Runs, float)
    for i in range(Runs):
        R,C,P,V,LR = ASEP(R,C,P,V,LR)
        Current[i] = float(sum(V)) / float(Runs)
        Vavg[i] = float(sum(V)) / float(len(V))
        Positions[i] = P+L*LR
    return Current, Vavg, Positions

        
```
