```python

def distribute1D(x,prob,N):
    b = prob
    F = cumsum(b)/sum(b)
    z = numpy.random.random_sample(N);
    sort(F)
    finv = x[searchsorted(F,z)]
    return finv
```

```python
x = arange(0.,20.1,.1)

N = 10000
z = numpy.random.random_sample(N)
hist(finv(z),50,normed=True,label="Sampled Distribution")
plot(x,f(x),label="True Distribution",lw=3,color='red')


plot(x,F(x),label="True Distribution",lw=3,color='red')
show()
```
